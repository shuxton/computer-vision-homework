### a) Human color perception

Explain how human color perception works, that is, how light of different frequencies (and mixtures of different frequencies) is perceived as different colors.

YOUR ANSWER HERE

1. Human color perception is based on the interaction of light with different wavelengths and the photoreceptor cells in the retina. Color is a characteristic of visible light, which is electromagnetic radiation with wavelengths ranging from approximately 400 to 700 nanometers. Short wavelengths (around 400–500 nm) are perceived as violet and blue, medium wavelengths (500–600 nm) as green and yellow, and long wavelengths (600–700 nm) as orange and red.

The retina contains two types of receptors: rods and cones. Rods are responsible for vision in low light conditions but do not perceive color. Cones are divided into three types, sensitive to short (blue), medium (green), and long (red) wavelengths. Each type of cone responds to a range of wavelengths rather than a single specific one. When light enters the eye, it stimulates certain cones depending on its wavelength and intensity. The resulting signals are transmitted through the optic nerve to the visual cortex in the brain, where they are processed and transformed into color perception.

If light consists of multiple wavelengths, the brain combines the signals to create the sensation of mixed colors. For example, a combination of red and green light is perceived as yellow, while red and blue light together appear purple. White is perceived when all three types of cones are stimulated equally. Color perception is also influenced by the surrounding context, lighting conditions, and individual differences in vision.

2. To create the perception of specific colors, you need to carefully select light sources with the right wavelengths or combinations of wavelengths. Here’s how it works for each color:

Orange

Orange is perceived when light has wavelengths 590-620 nm. A light source emitting in this range will appear orange. Alternatively, you can mix red and yellow light to achieve a similar result.

Brown

Brown is essentially a darker, less saturated version of orange. To create brown, use the same wavelengths as orange (590–620 nm) but reduce the brightness and place it against a darker background. The contrast between light and dark is crucial for perceiving brown.

Purple

Purple is not a spectral color—it’s created by mixing red and blue light. This can be achieved by combining long wavelengths (around 650–700 nm) from the red range with short wavelengths (around 400–450 nm) from the blue range. This combination stimulates the “blue” (S) and “red” (L) cones in the retina, resulting in the perception of purple.

White

White light is perceived when all three types of cones (S, M, L) are equally stimulated. You can achieve this by using light that covers the full spectrum (like sunlight) or by combining red, green, and blue light in equal proportions. Modern LED lights often create white light by blending these colors or using special phosphor coatings.

By carefully adjusting wavelengths and brightness, you can “trick” the eye into perceiving almost any color.

As a result, RGB is suitable for technical tasks where hardware is important, while HSV is better suited to situations where colour analysis or intuitive interaction is important. The choice depends on what the colour model is used for.

YOUR ANSWER HERE

Additive and subtractive color mixing are two fundamental principles used to create and manipulate colors in different contexts. Additive color mixing occurs when light of different wavelengths is combined, forming new colors through the addition of light. This model uses red, green, and blue (RGB) as primary colors. For example, red and green light create yellow, red and blue form magenta, green and blue produce cyan, and combining all three at full intensity results in white light. This principle is widely used in digital displays like televisions, computer monitors, and smartphones, as well as in stage lighting, where colored spotlights create effects. Additive mixing is crucial in digital imaging and simulations where light-based colors are required.

Subtractive color mixing, on the other hand, is used when pigments or dyes are combined. Each pigment absorbs certain wavelengths of light and reflects others. The primary colors in this model are cyan, magenta, and yellow (CMY). Cyan and magenta combine to produce blue, cyan and yellow create green, magenta and yellow result in red, and ideally, all three together should produce black (though in practice, the result is often dark brown or gray). Subtractive mixing is common in printing, where the CMYK (cyan, magenta, yellow, black) model ensures accurate color reproduction. It is also used in painting and dye manufacturing, where color adjustments rely on material absorption and reflection.

The key difference lies in their nature: additive mixing involves the combination of light, while subtractive mixing deals with the interaction of pigments or materials. Both models are essential for industries ranging from digital displays to physical printing, offering precise methods to reproduce and control color.

### a) Definition

Describe in your own words how convolution works. Also provide the formula from the lecture and explain its idea.

YOUR ANSWER HERE

Convolution is a mathematical operation that combines two functions to produce a third function, expressing how the shape of one function modifies the other. In image processing, convolution is typically used to apply a filter (or kernel) to an image to extract features, enhance details, or perform transformations like edge detection or blurring.

The process involves sliding the kernel (a small matrix) over the image, multiplying overlapping elements, and summing them to produce a new value for the corresponding pixel in the output image. This operation is repeated for each position of the kernel until the entire image is processed.


$$
g'(x, y) = \sum_{i=-m}^{m} \sum_{j=-n}^{n} k(i, j) \cdot g(x+i, y+j)
$$

The idea of convolution is to capture and enhance specific patterns or features in the image based on the kernel’s design. For instance:
1. Edge detection kernels emphasize differences in intensity by subtracting neighboring values, highlighting edges.
2.	Blurring kernels average the neighboring pixel values, smoothing the image.
3.	Sharpening kernels amplify differences between a pixel and its neighbors, enhancing details.

By changing the values and structure of the kernel, convolution can be adapted for a wide range of image processing tasks.

### c) Complexity

Assume an image $g$ of size $M\times N$ and a kernel $k$ of size $(2m+1)\times(2n+1)$. How many operations (additions and multiplications) are required to compute a convoluted image $g\ast k$ (of the same size as $g$)?

YOUR ANSWER HERE

1.	Number of Pixels:

The output image  $ g \ast k $ has the same size as the input image $ g $ , meaning there are  $ M \times N $ pixels in the output.

2.	Kernel Size:

The kernel  $ k $  has dimensions $ (2m+1) \times (2n+1) $, so it contains: $ K = (2m+1)(2n+1)$ elements.

3.	Operations per Pixel:

For each pixel in the output image:
	
The kernel is slid over the corresponding region of the input image.	

For each element of the kernel, one multiplication is performed with the corresponding pixel in the input image region.
	
All these products are then summed up to produce the final value for the output pixel.

Therefore, for each pixel, there are:

$ K = (2m+1)(2n+1) $ multiplications.

$ K-1 = (2m+1)(2n+1) - 1 $ additions.

4.	Total Operations:

Since the process is repeated for all  $ M \times N $ pixels, the total number of operations is:

Multiplications:

$ M \times N \times (2m+1)(2n+1)$ 

Additions:

$ M \times N \times \left((2m+1)(2n+1) - 1\right)$ 


5. Final Formula

Total Multiplications:

$ T_{\text{mult}} = M \cdot N \cdot (2m+1) \cdot (2n+1) $

Total Additions:

$ T_{\text{add}} = M \cdot N \cdot \left((2m+1)(2n+1) - 1\right) $



### d) Separability

What is a separable kernel? Describe, how it can be applied more efficiently. YOUR ANSWER HERE

A separable kernel is a convolution kernel that can be expressed as the outer product of two one-dimensional kernels. In mathematical terms, if the kernel  k  of size  m \times n  can be written as:
$ k(i, j) = k_1(i) \cdot k_2(j) $,
then the kernel is separable. This property allows us to replace the 2D convolution with two consecutive 1D convolutions: one along the rows and the other along the columns.

Using a separable kernel significantly reduces the number of operations required for convolution. Instead of directly convolving the kernel with the image in two dimensions, we can:
1.	First convolve the image with a 1D kernel $ k_1(i) $ along the rows.
2.	Then convolve the result with another 1D kernel $ k_2(j) $ along the columns.


Let the image size be $ M \times N $ and the kernel size be  m \times n .

Without Separability (Direct 2D Convolution)

For a single pixel:
We multiply $ m \times n $ elements of the kernel with the corresponding image region.

We sum up the $ m \times n $ results.

Thus, for each pixel, the number of operations (multiplications and additions) is $ m \times n - 1 $  additions and  m \times n  multiplications.

For the whole image:

Multiplications: $ M \cdot N \cdot m \cdot n $

Additions: $ M \cdot N \cdot (m \cdot n - 1) $

 With Separability 

1.	Row-wise 1D Convolution:
	•	Kernel $ k_1 $ has $ m $ elements.
	•	For a single pixel,  m  multiplications and $ m - 1 $ additions are required.
	•	For $ M \times N $ pixels: $ M \cdot N \cdot m $ multiplications and $ M \cdot N \cdot (m - 1) $ additions.
2.	Column-wise 1D Convolution:
	•	Kernel  $ k_2 $ has $ n $ elements.
	•	For a single pixel,  $ n $ multiplications and $ n - 1 $ additions are required.
	•	For $ M \times N $ pixels: $ M \cdot N \cdot n $ multiplications and $ M \cdot N \cdot (n - 1) $ additions.
3.	Total for Separable Kernel:
	•	Multiplications:  $M \cdot N \cdot (m + n) $
	•	Additions:  $M \cdot N \cdot (m + n - 2) $


Comparison

1.	Without Separability:
	•	Total operations: $ M \cdot N \cdot m \cdot n $ multiplications and $ M \cdot N \cdot (m \cdot n - 1) $ additions.
2.	With Separability:
	•	Total operations:  $M \cdot N \cdot (m + n) $ multiplications and  $M \cdot N \cdot (m + n - 2) $ additions.
3.	Reduction Factor:
The separable kernel reduces the number of operations by approximately:

$\text{Reduction Factor} = \frac{m \cdot n}{m + n}$

### d) Unsharp Mask

One method to sharpen images is Unsharp Mask in which a negative unsharp mask is added to the original image as follows:

$$\text{Sharpened Image} = \text{Original Image} + (\text{Original Image} - \text{Unsharp Image}) \cdot \text{Amount}$$

The unsharp image can be computed by convolution with a Gaussian Kernel. Implement unsharp masking with a $5\times5$ Gaussian Kernel and a sharpening amount of $1.5$. Use the allready defined gaussian kernel "gauss_5".

Hint: To get good results the final images needs to be clipped to values between $0$ and $255$, i.e. all negative values are set to zero and all values bigger than $255$ are set to $255$.

You may experiment with large or negative sharpening amounts.

* Why is Unsharp Masking sharpening an image?
* What is the difference between normalizing and clipping an image?


YOUR ANSWER HERE

* Why is Unsharp Masking sharpening an image?

Unsharp masking enhances edges and fine details in an image by emphasizing regions with high intensity differences. This is achieved by subtracting a blurred (unsharp) version of the image from the original image, isolating high-frequency components (edges). By amplifying these components and adding them back to the original image, the result appears sharper.

* What is the difference between normalizing and clipping an image?

Normalizing:

Scales the image values to a specific range, such as [0, 1] or [0, 255], while preserving proportionality between pixel intensities.

Often used for machine learning, histogram equalization, or to ensure uniform brightness levels.

Example: Dividing all pixel values by the maximum intensity.

Clipping:

Truncates pixel values that fall outside a specified range, setting negative values to 0 and values above the maximum (e.g., 255) to the maximum value.

Used to avoid overflows or invalid pixel intensities in image processing.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import convolve
from skimage import data

# Load an image
image = data.coins().astype(np.float32)

# Define sharpening amount
amount = 1.5

# Define the 5x5 Gaussian kernel
gauss_5 = 1/256 * np.asarray([[1, 4, 6, 4, 1],
                              [4, 16, 24, 16, 4],
                              [6, 24, 36, 24, 6],
                              [4, 16, 24, 16, 4],
                              [1, 4, 6, 4, 1]])

# Compute the unsharp (blurred) image
blurred_image = convolve(image, gauss_5)

# Apply the unsharp mask formula
unsharp_mask = image - blurred_image
sharpened_image = image + (unsharp_mask * amount)

# Clip the values to the range [0, 255]
sharpened_image = np.clip(sharpened_image, 0, 255)

# Display original and sharpened images
fig = plt.figure(figsize=(15, 7))

# Original image
a = fig.add_subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.axis('off')
a.set_title("Original Image")

# Sharpened image
a = fig.add_subplot(1, 2, 2)
plt.imshow(sharpened_image, cmap='gray')
plt.axis('off')
a.set_title("Sharpened Image")

plt.show()
```

## Assignment 4: Implementing Convolution (4 points)

Now implement your own 2-dimensional convolution function. The function should take an image and a kernel as arguments and return an image of the same size, containing the result of convolving the image with the kernel.

**(a)** First: you may notice a problem at the boundaries of the image. Describe the problem and possible solutions. Implement at least one of them in (b).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio.v3 as iio


def my_convolve2d(img, kern):
   
    # Get dimensions of the image and kernel
    M, N = img.shape  # Image dimensions
    m, n = kern.shape  # Kernel dimensions
    mh, nh = m // 2, n // 2  # Half kernel size for padding
    
    # Initialize the output image
    result = np.zeros((M, N), dtype=np.float32)
    
    # Pad the image with zeros
    padded_img = np.pad(img, ((mh, mh), (nh, nh)), mode='constant', constant_values=0)
    
    # Perform convolution
    for i in range(M):
        for j in range(N):
            # Extract the region of interest (ROI)
            roi = padded_img[i:i + m, j:j + n]
            # Element-wise multiplication and summation
            result[i, j] = np.sum(roi * kern)
    
    return result


# Load an image
image = iio.imread("imageio:coins.png").astype(np.float32)

# Define a kernel (e.g., edge detection)
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

# Apply custom convolution
convolved_image = my_convolve2d(image, kernel)

# Display the results
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(image, cmap='gray')
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(convolved_image, cmap='gray')
axes[1].set_title("Convolved Image")
axes[1].axis('off')

plt.show()
```

During convolution at the boundaries of an image, the kernel often extends beyond the edges, leading to undefined regions where the kernel overlaps with non-existent pixels. This causes several issues. First, there is a lack of valid data for performing convolution near the edges, resulting in incomplete or distorted output. Second, if these undefined regions are not handled properly, the resulting image may exhibit artifacts such as unnatural values or patterns along the borders. Finally, if boundary pixels are ignored, the size of the output image becomes smaller than the original, which is often undesirable in image processing tasks.

**(b)** Now implement your function. Apply it with different kernels and compare the results with the output of the library function `scipy.ndimage.correlate` (bonus question: why not `scipy.ndimage.convolve`?).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio.v3 as iio
import scipy.ndimage
import time  

def my_convolve2d(img, kern):
    # Get the dimensions of the image and the kernel
    M, N = img.shape  # Image dimensions
    m, n = kern.shape  # Kernel dimensions
    mh, nh = m // 2, n // 2  # Half of the kernel size for padding

    # Initialize the output image
    result = np.zeros((M, N), dtype=np.float32)

    # Pad the image with zeros
    padded_img = np.pad(img, ((mh, mh), (nh, nh)), mode='constant', constant_values=0)

    # Perform convolution
    for i in range(M):
        for j in range(N):
            # Extract the region of interest (ROI)
            roi = padded_img[i:i + m, j:j + n]
            # Element-wise multiplication and summation
            result[i, j] = np.sum(roi * kern)

    return result

# Apply the function to an image
image = iio.imread("imageio:coins.png").astype(np.float32)

# Different filters (kernels)
kernel = 1/4 * np.array([[1, 0, -1],
                         [2, 0, -2],
                         [1, 0, -1]])

# Measure time for the custom function
start_time = time.time()
my_result = my_convolve2d(image, kernel)
my_time = time.time() - start_time
print(f"Execution time of the custom function: {my_time:.4f} seconds")

# Measure time for the built-in function
start_time = time.time()
lib_result = scipy.ndimage.correlate(image, kernel)
lib_time = time.time() - start_time
print(f"Execution time of the built-in function: {lib_time:.4f} seconds")

# Visualize the results
fig = plt.figure(figsize=(15, 7))

fig.add_subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

fig.add_subplot(1, 3, 2)
plt.title(f"Custom Implementation ({my_time:.4f} s)")
plt.imshow(my_result, cmap='gray')
plt.axis('off')

fig.add_subplot(1, 3, 3)
plt.title(f"Library Implementation ({lib_time:.4f} s)")
plt.imshow(lib_result, cmap='gray')
plt.axis('off')

plt.show()
```
